```python
import face_recognition
import cv2
import numpy as np
import os
#import pandas as pd

gaurav_image = face_recognition.load_image_file(r"Downloads//ML//known_gaurav//gaurav.jpg")
gaurav_face_encoding = face_recognition.face_encodings(gaurav_image)[0]
print("gaurav")

# Create arrays of known face encodings and their names
known_face_encodings = [
    gaurav_face_encoding,
]

known_face_names = [
    "gaurav"
]

# Dictionary to map unknown image filenames to their actual names (ground truth)
ground_truth = {
   "gaurav": "gaurav.jpg" ,
}

```

```python
#import os
# Initialize some variables

face_locations = []
face_encodings = []
face_names = []

# Process unknown images
path = r"Downloads//ML//unkown_gaurav"
correct_predictions = 0
total_predictions = 0


for (root, dirs, files) in os.walk(path, topdown=True):
    for file in files:
        unknown_image_path = os.path.join(root, file)  # Use os.path.join here
        if not os.path.exists(unknown_image_path):
            print(f"File not found: {unknown_image_path}")
            continue



# for (root, dirs, files) in os.walk(path, topdown=True):
#     for file in files:
        unknown_image = face_recognition.load_image_file(unknown_image_path)  # Correct path usage
        face_locations = face_recognition.face_locations(unknown_image)
        face_encodings = face_recognition.face_encodings(unknown_image, face_locations)
        face_names = []
        for face_encoding in face_encodings:
            matches = face_recognition.compare_faces(known_face_encodings, face_encoding)
            name = "Unknown"
            face_distances = face_recognition.face_distance(known_face_encodings, face_encoding)
            best_match_index = np.argmin(face_distances)
            if matches[best_match_index]:
                name = known_face_names[best_match_index]
                
            face_names.append(name)
            print(f"File: {file}, Predicted Name: {name}")
            file_key = file.split(' ')[0] 
            db_filename=ground_truth[file_key]
            name_db_filename=db_filename.split(".")[0]
            
            
            
            if True:
                print(name)
                print( name_db_filename)
                if name == name_db_filename :
                    correct_predictions += 1

                    
                    #print(correct_predictions)
                    total_predictions += 1
                else:
                    total_predictions+=1
            print("running percentage=",(correct_predictions/total_predictions)*100)
            

# Calculate and print the accuracy
if total_predictions > 0:
    accuracy = (correct_predictions / total_predictions) * 100
    print(f"Accuracy: {accuracy:.2f}%")
else:
    print("No predictions were made.")
```
